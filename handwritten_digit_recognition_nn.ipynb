```python
import numpy as np
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
# importing the dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# checking shapes
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# checking features of the first sample
x_train[0]
```

```python
# shape of the first sample
x_train[0].shape
```

```python
# plotting the first sample
plt.matshow(x_train[0])
```

```python
y_train[:5]
```

```python
# scalling the features
x_train_scaled = x_train/255
x_test_scaled = x_test/255
```

```python
x_train_scaled[0]
```

```python
# flattening the 28x28 images to a single dimensional array of 784 elements
x_train_flattened = x_train_scaled.reshape(len(x_train_scaled), 28*28)
x_test_flattened = x_test_scaled.reshape(len(x_test_scaled), 28*28)
```

```python
# shapes of x_train and x_test after flattening the image dimensions
x_train_flattened.shape, x_test_flattened.shape
```

```python
x_train_flattened[0]
```

```python
# creating a NN model
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

# compiling the model
model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# fitting the model
model.fit(x_train_flattened, y_train, epochs=10)
```

```python
# model evaluation using the test data set
model.evaluate(x_test_flattened, y_test)
```

```python
# plotting the first test sample
plt.matshow(x_test[0])
```

```python
# predicting accuracy

y_predicted = model.predict(x_test_flattened)
y_predicted[0]
```

```python
# checking the maximum prediction
print("Actual\tPredicted\tAccuracy")
for i in range(0, len(x_test_flattened)):
    actual_vals = y_test[i]
    predicted_vals = np.argmax(y_predicted[i])
    print(actual_vals, "\t", predicted_vals)
```

```python
# y_predicted labels
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
# confusion matrix
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
# visualizing the cm
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.title("Model 1 Confusion Matrix")
plt.xlabel('Predicted value')
plt.ylabel('Actual value')
plt.show()
```

```python
# same model but now with a hidden layer

# creating the model
model2 = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

# compiling the model
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# fitting the model
model2.fit(x_train_flattened, y_train, epochs=10)
```

```python
# evaluating new model
model2.evaluate(x_test_flattened, y_test)
```

```python
# y_predicted labels
y_predicted2 = model2.predict(x_test_flattened)
y_predicted_labels2 = [np.argmax(i) for i in y_predicted2]
y_predicted_labels[:5]
```

```python
# confusion matrix for model 2
cm2 = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels2)
cm2
```

```python
# visualizing the cm for model 2
plt.figure(figsize=(10,7))
sn.heatmap(cm2, annot=True, fmt='d')
plt.title("Model 2 Confusion Matrix")
plt.xlabel('Predicted value')
plt.ylabel('Actual value')
plt.show()
```

```python
# third model using keras flattened layer

# creating model 3
model3 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(500, activation='sigmoid'),
    keras.layers.Dense(300, activation='sigmoid'),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(10, activation='sigmoid')
])

# compiling model 3
model3.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# fitting model 3
model3.fit(x_train, y_train, epochs=10)
```
